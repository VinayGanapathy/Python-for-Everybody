```python
# Loops and iterations
```

```python
n = 5
while n > 0 :
  print(n)
  n = n - 1
print('Blastoff!')
print(n)    # variable 'n' is called the iteration variable
```

```python
# Example of a badly constructed loop
n = 5
while n > 0 :
  print('Lather')
  print('Rinse')
print('Dry off!') # This loop run infinitely long. It never exits loop.
```

```python
# Example of a badly constructed loop
n = 0
while n > 0 :
  print('Lather')
  print('Rinse')
print('Dry off!') # This loop is called zero trip loop
```

```python
# General rule with while loop is
#1. the iteration variable is defined before the loop begins.(outside the loop)
#2. Iteration variable is modified within the loop.
```

```python
while True:
  line = input ('> ')
  if line == 'done':
    break
  print(line)
print('Done!')           # 'break' statement will get out of the loop. It exits the innermost loop.
```

```python
while True:
  line = input ('> ')
  if line[0] == '#':
    continue
  if line == 'done':
    break
  print(line)
print('Done!')     # 'continue' redirects to the top of the loop. The loop continues.
```

```python
# Loop idioms (what we do in loops)
```

```python
print('Before')
for thing in [9,41,12,3,74,15]:
  print(thing)
print('After')
```

```python
# finding the largest number in a list

largest_so_far = -1
print('Before ', largest_so_far)
for the_num in [9,41,12,3,74,15]:
  if the_num > largest_so_far:
    largest_so_far = the_num
  print(largest_so_far, the_num)

print('After', largest_so_far)
```

```python
# More loop patterns
```

```python
# Counting in a loop
zork = 0
print('Before', zork)
for thing in [9,41,12,3,74,15]:
  zork = zork + 1
  print(zork, thing)
print('After', zork)
```

```python
 # Calculating the total of the elements in a loop (cumulative total)
zork = 0
print('Before', zork)
for thing in [9,41,12,3,74,15]:
  zork = zork + thing
  print(zork, thing)
print('After', zork)
```

```python
 # Calculating the average of the elements in a loop
count = 0
sum = 0
print('Before', count, sum)
for value in [9,41,12,3,74,15]:
  count = count + 1
  sum = sum + value
  print(count, sum, value)
print('After', count, sum, sum/count)
```

```python
 # Filtering in a loop

print('Before')
for value in [9,41,12,3,74,15]:
  if value > 20 :
    print('Large number', value)
print('After')
```

```python
# Search Using a Boolean variable

found = False
print('Before', found)
for value in [9,41,12,3,74,15]:
  if value == 3:
    found = True
  print(found, value)
print('After', found)
```

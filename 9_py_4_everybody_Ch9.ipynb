```python
# code to build the histogram of names
counts = dict()
names = ['csev', 'cwen', 'csev', 'zqian','cwen']
for name in names:
  if name not in counts:
    counts[name] = 1
  else:
    counts[name] = counts[name] + 1
print(counts)
```

```python
# using 'get' method to achieve the histogram
counts = dict()
names = ['csev', 'cwen', 'csev', 'zqian','cwen']
for name in names:
    counts[name] = counts.get(name, 0) + 1    # '0' is a default value to the name
print(counts)
```

```python
# counting words in a text
counts = dict()
print('Enter a line of text: ')
line = input('')
words = line.split()
print('Words: ', words)
print('Counting')
for word in words:
  counts[word] = counts.get(word, 0) + 1
print('Counts', counts)
```

```python
# Comments always starts with a # sign. An example is shown below.


# Get the name of the file and open it
name = input("enter file : ")
handle = open(name, 'r')

#Count word frequency
counts = dict()
for line in handle:
  words = line.split()
  for word in words:
    counts[word] = counts.get(word,0)+1

# Find the most common word
bigcount = None
bigword = None
for word,count in counts.items():
  if bigcount is None or count > bigcount:
    bigword = word
    bigcount = count

# all done
print(bigword,bigcount)
```

```python
fname = input('Enter File: ')
if len(fname) < 1 : fname = '/content/drive/MyDrive/Colab Notebooks/clown.txt'
hand = open(fname)

di = dict()
for line in hand:
  line = line.rstrip()
  #print(line)
  wds = line.split()
  #print(wds)
  for w in wds:
    #print('**', w, di.get(w,-99))
    oldcount = di.get(w,0)     # if the key is not there the count is zero
    print(w,'old', oldcount)
    newcount = oldcount + 1
    di[w] = newcount
    print(w,'new', newcount)
    #print(w, di[w])
print(di)
```

```python
fname = input('Enter File: ')
if len(fname) < 1 : fname = '/content/drive/MyDrive/Colab Notebooks/clown.txt'
hand = open(fname)

di = dict()
for line in hand:
  line = line.rstrip()
  wds = line.split()
  for w in wds:
    # idiom : retrieve/create/update counter
    di[w] = di.get(w,0) + 1

print(di)


# finding the most common word
largest = -1
theword = None
for k,v in di.items():
  #print(k,v)
  if v > largest:
    largest = v
    theword = k    # capture and remember the word that was largest

print(theword, largest)
```

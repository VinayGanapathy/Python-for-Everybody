```python
# Matching and extracting
```

```python
# Extracting numerics from the text
import re
x = 'My 2 favorite numbers are 19 and 42'
y = re.findall('[0-9]+', x)
print(y)
```

```python
# Extracting one or more uppercase (vowels/letters) from the text
import re
x = 'My 2 favorite numbers are 19 and 42'
y = re.findall('[AEIOU]+', x)
print(y)
```

```python
# Greedy and non greedy matching
# Greedy matching
import re
x = 'From: Using the : character'
y = re.findall('^F.+:', x)
print(y)
```

```python
# Non - Greedy matching
import re
x = 'From: Using the : character'
y = re.findall('^F.+?:', x)
print(y)
```

```python
data = 'From stephan.marquard@uct.ac.za Sat Jan  5  09:14:16 2008'
y = re.findall('\S+@\S+', data)
print(y)
```

```python
data = 'From stephan.marquard@uct.ac.za Sat Jan  5  09:14:16 2008'
y = re.findall('^From (\S+@\S+)', data)
print(y)
```

```python
data = 'From stephan.marquard@uct.ac.za Sat Jan  5  09:14:16 2008'
y = re.findall('^From \S+@\S+', data)
print(y)
```

```python
# Parsing examples - String parsing
```

```python
# Old example (without using re)

data = 'From stephan.marquard@uct.ac.za Sat Jan  5  09:14:16 2008'
atpos = data.find('@')
print(atpos)

sppos = data.find(' ', atpos)
print(sppos)

host = data[atpos+1:sppos]
print(host)
```

```python
# other way of the same example
data = 'From stephan.marquard@uct.ac.za Sat Jan  5  09:14:16 2008'
words = data.split()
email = words[1]
pieces = email.split('@')
print(pieces[1])
```

```python
# using re
import re
data = 'From stephan.marquard@uct.ac.za Sat Jan  5  09:14:16 2008'
y = re.findall('@([^ ]*)', data)
print(y)
```

```python
import re
data = 'From stephan.marquard@uct.ac.za Sat Jan  5  09:14:16 2008'
y = re.findall('^From .*@([^ ]*)', data)
print(y)
```

```python
# Finding maximum spam confidence

import re

hand = open('/content/drive/MyDrive/Colab Notebooks/mbox-short.txt')
numlist = list()
for line in hand:
  line = line.rstrip()
  stuff = re.findall('X-DSPAM-Confidence: ([0-9.]+)', line)
  if len(stuff) != 1 : continue
  num = float(stuff[0])
  numlist.append(num)
print('Maximum:', max(numlist))
```

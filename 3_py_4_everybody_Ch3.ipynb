```python
x = 5
if x == 5:
  print('Equals 5')
if x > 4 :
  print('Greater than 4')
if x >= 5 :
  print('Greater than or equal to 5')
if x < 6 :
  print("less than 6")
if x <= 5:
  print('less than or equal to 5')
if x != 6 :
  print('Not equal to 6')
```

```python
# Indentation is used when there is a 'if statement' block (4 spaces is the general norm of indentation but generally editors automatically have this indentation)
# avoid using 'Tab' for the indentation
```

```python
x = 5
if x > 2:
  print('Bigger than 2')   # identation starts here and remains for the whole 'if' block
  print('Still bigger') 
print('Done with 2')   # Indentation ends here. so till the top line it comes in the 'if' block

for i in range(5):
  print(i)
  if i > 2:
    print('Bigger than 2')
  print('Done with i', i)
print('All done')  # An example of nested loops and indentation
```

```python
# Try and except (avoids Traceback)

astr = 'Hello Bob'
istr = int(astr)
print('First', istr)
astr = '123'
istr = int(astr)
print('Second', istr)
```

```python
try:
  astr = 'Hello Bob'
  istr = int(astr)
  print('First', istr)
except:
  astr = '123'
  istr = int(astr)
  print('Second', istr)
```

```python
astr = ' Hello Bob'
try:
  istr = int(astr)
except:
  istr = -1

print('First', istr)

astr = '123'
try:
  istr = int(astr)
except:
  istr = -1
print('Second', istr)
```

```python
astr = 'Bob'
try:
  print('Hello')
  istr = int(astr)
  print('There')
except:
  istr = -1

print('Done', istr)
```

```python
astr = 'Bob'
try:
  print('Hello')
  print('till here')
  istr = int(astr)
  print('There')
except:
  istr = -1

print('Done', istr)
```
